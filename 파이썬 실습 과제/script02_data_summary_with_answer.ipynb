## 데이터 요약

### 라이브러리와 데이터읽기

```python
import pandas as pd
data = pd.read_csv('data01_iris.csv')
```

### 단변량 데이터 요약

```python
# 자세한 내용은 pandas 섹션의 데이터 정보와 시각화 참조
# data.info()
# data.describe()
# data['Species'].value_counts()
# data.hist()
# data.boxplot()
```

###이변량 데이터 요약 (수치-수치)

```python
data
```

```python
# 상관계수
data[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].corr()
```

```python
# 산점도
data.plot.scatter('Sepal.Length','Sepal.Width')
```

### 이변량 데이터 요약 (수치-범주)

```python
# 범주별 박스 플롯
data.boxplot('Sepal.Length',by='Species')
```

```python
# 범주별 평균
data.groupby('Species').mean()
```

```python
# 범주별 분산
data.groupby('Species').var()
```

```python
# setosa와 versicolor 사이의 SMD
import math

ms = data.groupby('Species').mean()
vs = data.groupby('Species').var()
ns = data['Species'].value_counts()

m1 = ms['Sepal.Length']['setosa']
m2 = ms['Sepal.Length']['versicolor']
n1 = ns['setosa']
n2 = ns['versicolor']
v1 = vs['Sepal.Length']['setosa']
v2 = vs['Sepal.Length']['versicolor']
d = abs(m1-m2)
s = math.sqrt( ((n1-1)*v1 + (n2-1)*v2)/(n1+n2-2) )
smd = d/s
print(smd)
```

### 이변량 데이터 분석 (범주-범주)

```python
# 수치형 변수를 범주형으로 변환
sl_bin = pd.cut(data['Sepal.Length'],bins=3,labels=['small','middle','large'])
sl_bin
```

```python
sl_bin.value_counts()
```

```python
# data에 새로운 변수로 추가
data['SL.bin'] = sl_bin
data.head()
```

```python
# 크로스테이블
pd.crosstab(data['Species'],data['SL.bin'])
```

3. iris 데이터셋에서 각 품종 별로 색을 다르게하여 Sepal Length와 Sepal Width 사이의 scatter plot을 그리시오.



```python
data_setosa = data[data['Species']=='setosa']
data_versicolor = data[data['Species']=='versicolor']
data_virginica = data[data['Species']=='virginica']

plt.figure(figsize=(8,6))
plt.scatter(data_setosa['Sepal.Length'], data_setosa['Sepal.Width'], label = 'Setosa')
plt.scatter(data_versicolor['Sepal.Length'], data_versicolor['Sepal.Width'], label = 'Versicolor')
plt.scatter(data_virginica['Sepal.Length'], data_virginica['Sepal.Width'], label = 'Virginica')

plt.xlabel('Sepal Length')
plt.ylabel('Speal Width')
plt.legend()
plt.show()
```
