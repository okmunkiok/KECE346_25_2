1. 다음의 코드는 각 학생의 영어, 수학, 과학 성적을 표현하였다. Jane의 수학 성적과 John을 과학 성적의 합을 출력하는 코드를 작성하시오.

```python
Scores = {
    'Kim': [95,68,77],
    'Jane': [89,75,92],
    'John': [81,99,97],
    'Tom': [77,85,78]
}
# put your code here
janes_math_score = Scores['Jane'][1]
johns_science_score = Scores['John'][2]

total = janes_math_score + johns_science_score
print("Total:", total)
```

2. 다음의 코드는 각 학생의 영어, 수학, 과학 성적을 표현하였다. 각 학생의 수학점수를 A/B/C/F로 평가하고 출력하시오. 90이상은 A, 80점 이상은 B, 70점 이상은 C, 그 이하는 F로 평가하시오.  

```python
Scores = {
    'Kim': [95,68,77],
    'Jane': [89,75,92],
    'John': [81,99,97],
    'Tom': [77,85,78]
}
# put your code here
for name, scores in Scores.items():
    Name = name
    Score = scores[1]

    if Score > 90:
        grade = 'A'
    elif Score > 80:
       grade = 'B'
    elif Score > 70:
        grade = 'C'
    else:
        grade = 'F'
    print(f"{Name}'s grade: {grade}")
```

3. 위의 문제에서 수학 성적의 학점 평균을 구하시오. A는 4.0, B는 3.0, C는 2.0, D는 1.0, F는 0.0 으로 평가하시오.

```python
gpa_mapping = {'A': 4.0, 'B': 3.0, 'C': 2.0, 'D': 1.0, 'F': 0.0}
total = 0
people = 0

for name, scores in Scores.items():
    Name = name
    Score = scores[1]

    if Score > 90:
        grade = 'A'
    elif Score > 80:
       grade = 'B'
    elif Score > 70:
        grade = 'C'
    else:
        grade = 'F'
    gpa = gpa_mapping[grade]
    total += gpa
    people += 1

average = total/people
print(average)
```

4. 1000이하의 소수의 개수를 구하는 코드를 작성하시오.

```python
# 소수의 개수를 셀 변수
count = 0

for num in range(1, 1001):
    # 약수의 개수를 셀 변수
    n=0
    for k in range(1,num+1):
      if num%k == 0:
          n += 1
    if n==2: # 약수의 개수가 2개면 소수
      count += 1

print("1000 이하의 소수의 개수:", count)
```
