## 데이터 요약

### 라이브러리와 데이터읽기

```python
import pandas as pd
data = pd.read_csv('data01_iris.csv')
```

### 단변량 데이터 요약

```python
# 자세한 내용은 pandas 섹션의 데이터 정보와 시각화 참조
# data.info()
# data.describe()
# data['Species'].value_counts()
# data.hist()
# data.boxplot()
```

###이변량 데이터 요약 (수치-수치)

```python
data
```

```python
# 상관계수
data[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].corr()
```

```python
# 산점도
data.plot.scatter('Sepal.Length','Sepal.Width')
```

### 이변량 데이터 요약 (수치-범주)

```python
# 범주별 박스 플롯
data.boxplot('Sepal.Length',by='Species')
```

```python
# 범주별 평균
data.groupby('Species').mean()
```

```python
# 범주별 분산
data.groupby('Species').var()
```

```python
# setosa와 versicolor 사이의 SMD
import math

ms = data.groupby('Species').mean()
vs = data.groupby('Species').var()
ns = data['Species'].value_counts()

m1 = ms['Sepal.Length']['setosa']
m2 = ms['Sepal.Length']['versicolor']
n1 = ns['setosa']
n2 = ns['versicolor']
v1 = vs['Sepal.Length']['setosa']
v2 = vs['Sepal.Length']['versicolor']
d = abs(m1-m2)
s = math.sqrt( ((n1-1)*v1 + (n2-1)*v2)/(n1+n2-2) )
smd = d/s
print(smd)
```

### 이변량 데이터 분석 (범주-범주)

```python
# 수치형 변수를 범주형으로 변환
sl_bin = pd.cut(data['Sepal.Length'],bins=3,labels=['small','middle','large'])
sl_bin
```

```python
sl_bin.value_counts()
```

```python
# data에 새로운 변수로 추가
data['SL.bin'] = sl_bin
data.head()
```

```python
# 크로스테이블
pd.crosstab(data['Species'],data['SL.bin'])
```

4. carseat 데이터셋의 Sales에 대하여 요약을 수행하고, 박스플롯과 히스토그램을 그리시오. SalesBin을 요약하시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

car_data = pd.read_csv('data03_carseat.csv')

print("--- 1. Sales 컬럼 기술 통계 요약 ---")
print(car_data['Sales'].describe())
print("\n" + "="*50 + "\n") # 줄바꿈 및 구분선

print("--- 2. Sales 컬럼 박스플롯 ---")
print(car_data.boxplot(column='Sales'))


print("--- 3. Sales 컬럼 히스토그램 ---")
print(car_data.hist(column='Sales'))
```

```python
print("--- 4. SalesBin 컬럼 요약 ---")
print(car_data['SalesBin'].describe())
```

5. carseat 데이터셋의 다음의 변수에 대해 이변량 분석을 수행하시오.<br>
(1) Sales-CompPrice<br>
(2) Sales-Urban<br>
(3) SalesBin-Urban<br>

```python
print(car_data['Sales'].dtypes)
print(car_data['CompPrice'].dtypes)
print(car_data['Urban'].dtypes)
print(car_data['SalesBin'].dtypes)
# print(car_data)

print("---- (1) Sales-CompPrice (수치 - 수치) ----")
print(car_data[['Sales','CompPrice']].corr())
print(car_data.plot.scatter('Sales','CompPrice'))

print("\n\n---- (2) Slaes-Urban (수치 - 범주) ----")
print(car_data.boxplot('Sales', by='Urban'))
```

```python
print(car_data.groupby('Urban')['Sales'].mean())
print(car_data.groupby('Urban')['Sales'].var())

print("\n\n---- Urban과 Urban x 사이의 SMD ----")
import math

ms = car_data.groupby('Urban')['Sales'].mean()
vs = car_data.groupby('Urban')['Sales'].var()
ns = car_data['Urban'].value_counts()

m1 = ms['No']
m2 = ms['Yes']
n1 = ns['No']
n2 = ns['Yes']
v1 = vs['No']
v2 = vs['Yes']

d = abs(m1-m2)
s = math.sqrt( ((n1-1)*v1 + (n2-1)*v2)/(n1+n2-2) )
smd = d/s
print(smd)
```

```python
print("\n\n---- (3) SlaesBin-Urban (범주 - 범주) ----")
pd.crosstab(car_data['SalesBin'],car_data['Urban'])
```
