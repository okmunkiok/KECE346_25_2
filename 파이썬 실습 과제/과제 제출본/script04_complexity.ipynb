# 모델 복잡성

## KNN 예제

### 라이브러리와 데이터

```python
# 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 읽기
data = pd.read_csv('data02_iris_noisy.csv')
X = data[ ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'] ]
y = data['Species']
print( X.shape, y.shape )
data.head()
```

```python
# 훈련/평가 데이터의 분리
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)
```

### KNN 모델 훈련

```python
# 모델 설정
from sklearn.neighbors import KNeighborsClassifier
f = KNeighborsClassifier(n_neighbors=5)
```

```python
# 모델 훈련
f.fit(xtrain, ytrain)
```

```python
# 모델을 이용한 예측
yhat_test = f.predict(xtest)
print(yhat_test)
```

```python
# 모델 평가: 정확도 측정
f.score(xtest,ytest)
```

### 파라메터에 변화에 따른 정확도 변화

```python
# K를 1에서 50까지 변화시키면서 훈련/평가 데이터의 정확도를 측정
klist = np.arange(1,51)
train_acc = []
test_acc = []
for k in klist:
  f = KNeighborsClassifier(n_neighbors=k)
  f.fit(xtrain,ytrain)
  train_acc.append( f.score(xtrain,ytrain) )
  test_acc.append( f.score(xtest,ytest) )
```

```python
# 그래프로 표현
plt.plot(klist,train_acc,'b-',klist,test_acc,'r-')
plt.legend(labels=['Train','Test'])
plt.show()
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# KNeighborsClassifier 대신 KNeighborsRegressor를 불러옵니다.
from sklearn.neighbors import KNeighborsRegressor

# (1) 데이터 읽기 및 분리
# 데이터를 읽어옵니다.
diabetes_data = pd.read_csv('data06_diabetes.csv')

# 입력 변수(X)와 목표 변수(Y)를 분리합니다.
X_diabetes = diabetes_data.drop('Y', axis=1)
y_diabetes = diabetes_data['Y']

# 훈련 데이터와 평가 데이터로 분리합니다. (평가 데이터 비율 40%, random_state=1)
xtrain_diabetes, xtest_diabetes, ytrain_diabetes, ytest_diabetes = train_test_split(
    X_diabetes, y_diabetes, test_size=0.4, random_state=1
)

print("--- 훈련/평가 데이터 형태 ---")
print("xtrain:", xtrain_diabetes.shape)
print("xtest:", xtest_diabetes.shape)
print("ytrain:", ytrain_diabetes.shape)
print("ytest:", ytest_diabetes.shape)
print("-" * 25)

# (2) K=5일 때 회귀 모델 훈련 및 평가
# KNeighborsRegressor를 사용합니다.
knn_reg_model = KNeighborsRegressor(n_neighbors=5)

# 훈련 데이터를 이용하여 모델을 훈련합니다.
knn_reg_model.fit(xtrain_diabetes, ytrain_diabetes)

# 평가 데이터에서 모델의 성능(R-squared)을 측정합니다.
r2_score_k5 = knn_reg_model.score(xtest_diabetes, ytest_diabetes)
print(f"K=5일 때 평가 데이터 R-squared: {r2_score_k5:.4f}")
print("-" * 25)


# (3) K 값 변화에 따른 성능 측정 및 시각화
k_list = np.arange(1, 101)
train_scores = []
test_scores = []

for k in k_list:
    # 각 K값에 대한 회귀 모델을 생성하고 훈련합니다.
    temp_model = KNeighborsRegressor(n_neighbors=k)
    temp_model.fit(xtrain_diabetes, ytrain_diabetes)

    # 훈련 데이터와 평가 데이터의 R-squared를 계산하여 각 리스트에 추가합니다.
    train_scores.append(temp_model.score(xtrain_diabetes, ytrain_diabetes))
    test_scores.append(temp_model.score(xtest_diabetes, ytest_diabetes))

# K 값의 변화에 따른 R-squared 변화를 그래프로 시각화합니다.
plt.figure(figsize=(10, 6))
plt.plot(k_list, train_scores, 'b-', label='Train R-squared')
plt.plot(k_list, test_scores, 'r-', label='Test R-squared')
plt.title('KNN Regressor: R-squared vs. Number of Neighbors (K)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('R-squared (결정계수)')
plt.legend()
plt.grid(True)
plt.show()
```
